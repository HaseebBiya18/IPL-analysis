```python
#who is overall orange cap holder
bnames=df["batsman"].unique()
data={}
for i in bnames:
    single_batsman_run=df[df["batsman"]==i]["batsman_runs"].sum()
    data[i]=single_batsman_run
print(data)
    
```

```python
#mx, bat_mx are functions created to store data . (f {}) is used to 
mx=0
bat_mx=""
for i in data:
    if data[i]>mx:
        mx=data[i]
        bat_mx=i
print(f"batsman name is = {bat_mx} and the runs he scored is {mx}")
```

```python
#to find total wickets taken by YS Chahal
total_wickets = df[df["bowler"] == "YS Chahal"]["player_dismissed"].dropna().count()
print("Total wickets taken by YS Chahal:", total_wickets)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("Match.csv")
# Calculate total runs per over
total_runs_per_over = df.groupby('over')['total_runs'].sum()
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(total_runs_per_over.index, total_runs_per_over.values, marker='o', linestyle='-')
plt.title('Total Runs per Over')
plt.xlabel('Over')
plt.ylabel('Total Runs')
plt.grid(True)
plt.show()
```

```python
#countplot
plt.figure(figsize=(10, 6))
sns.countplot(x='batting_team', data=df)
plt.xticks(rotation=45)
plt.title('Matches by Batting Team')
plt.show()
```

```python
#Histogram
plt.hist(df['batsman_runs'])
plt.title('Distribution of Batsman Runs')
plt.xlabel('Runs')
plt.ylabel('Frequency')
plt.show()
```

```python
#boxplot
sns.boxplot(x='inning', y='batsman_runs', data=df)
plt.title('Batsman Runs by Inning')
plt.show()
```

```python
#Scatter plot
plt.scatter(df['over'], df['batsman_runs'])
plt.title('Batsman Runs by Over')
plt.xlabel('Over')
plt.ylabel('Runs')
plt.show()
```
